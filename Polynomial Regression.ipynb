```python
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model  import LinearRegression
from sklearn.metrics       import r2_score
import matplotlib.pyplot as plt
import numpy             as np
import pandas            as pd
plt.style.use('dark-style.mplstyle')




# Generate X and Y
X = pd.DataFrame({
    'A': [i for i in range(1,101)],
    'B': [i for i in range(2,102)]
})

Y = pd.DataFrame()
Y['C'] = X['A']**2 + X['A']

display( pd.concat([X,Y], axis=1) )

```

<br>

# Sklearn

```python
# Do poly regression
poly   = PolynomialFeatures(degree=2)
X_poly = poly.fit_transform(X)
# Generate the model
model = LinearRegression()
model.fit(X_poly, Y)


# Display the result of poly features and he coeefs of model
display( pd.concat([X,Y,pd.DataFrame(X_poly)], axis=1) )
print(model.coef_)

Y_pred = model.predict(X_poly)
# Plot best fit line
# plt.scatter(X,Y, s=2, zorder=2)
# plt.plot(X,Y_pred, color='#bfbbd9', zorder=1)
# plt.show()
r2_score(Y, Y_pred)
```
